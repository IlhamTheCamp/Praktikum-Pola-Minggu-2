```python
#Data dalam grafik

traffic = np.genfromtxt(r"D:\KULIAH\Magister Ilkom IPB\Semester 2\Pengenalan Pola\Bahan Praktikum 2\Web_Traffic_Data.tsv",delimiter=',')
print(traffic[:10])
print(traffic.shape)
x = traffic[:,0]
y = traffic[:,1]
x = x[~np.isnan(y)]
y = y[~np.isnan(y)]

plt.scatter(x,y)
plt.title("Web traffic last month")
plt.xlabel("Time")
plt.ylabel("Hits/hour")
plt.xticks([w*7*24 for w in range(10)],['week %i' %w for w in
range(10)])
plt.autoscale(tight=True)
plt.grid()
```
